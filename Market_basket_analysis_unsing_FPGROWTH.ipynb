```python
from mlxtend.frequent_patterns import fpgrowth
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.preprocessing import TransactionEncoder
```

```python
d = pd.read_csv("/content/Market_Basket_Optimisation.csv")
d.head()
```

```python
te = TransactionEncoder()
te_ary = te.fit(d).transform(d)
d = pd.DataFrame(te_ary, columns=te.columns_)
d
```

```python
fpgrowth(d,min_support = 0.001,use_colnames = True)
```
